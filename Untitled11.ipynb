```python
import pandas as pd
import numpy as np
import matplotlib as mp
```

```python
df = pd.read_csv(r'/Users/macuser/Documents/Briths AL INternship/BA_reviews.csv')
df
df.head()
```

```python
unnamed_columns = [col for col in df.columns if 'Unnamed: 0' in col]
df = df.drop(columns=unnamed_columns)
df.reset_index(drop=True, inplace=True)

df
```

```python
pip install textblob
```

```python
from textblob import TextBlob
```

```python


# Create a new column for sentiment polarity
df['sentiment_polarity'] = df['reviews'].apply(lambda text: TextBlob(text).sentiment.polarity)

# Classify sentiment labels
df['sentiment_label'] = df['sentiment_polarity'].apply(lambda polarity: "Positive" if polarity > 0 else ("Negative" if polarity < 0 else "Neutral"))

# Display the DataFrame with sentiment analysis results
print(df)
```

```python
df
```

```python
sentiment_counts = df['sentiment_label'].value_counts()
sentiment_counts
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(df.index, df['sentiment_polarity'], color='blue', alpha=0.5)
plt.title('Sentiment Polarity of Reviews')
plt.xlabel('Review Index')
plt.ylabel('Sentiment Polarity')
plt.grid(True)
plt.show()
```

```python
labels = sentiment_counts.index
sizes = sentiment_counts.values
colors = ['green','red','yellow']
explode = (0.1,0,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Sentiment Distribution')
plt.show()
```
